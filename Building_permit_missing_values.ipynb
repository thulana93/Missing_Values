# 6. Handling missing values (imputation, deletion)

```python
# NaN replaced with 'N' for Follwing features (As those are having only Yes and No)
#  Voluntary Soft-Story Retrofit
#  Fire Only Permit
#  TIDF Compliance
#  Site Permit
#  Structural Notification
dataset[['Voluntary Soft-Story Retrofit','Fire Only Permit','TIDF Compliance','Site Permit','Structural Notification']] = dataset[['Voluntary Soft-Story Retrofit','Fire Only Permit','TIDF Compliance','Site Permit','Structural Notification']].replace(np.nan, 'N')
# dataset['Voluntary Soft-Story Retrofit'] = dataset['Voluntary Soft-Story Retrofit'].replace(np.nan, 'N')
dataset.isnull().sum()
```

```python
# Following values are replace with mean value
    # Number of Proposed Stories
    # Estimated Cost
    # Revised Cost
    # Plansets
    # Existing Units

dataset['Number of Proposed Stories'].fillna(value=dataset['Number of Proposed Stories'].mean(), inplace=True)
dataset['Estimated Cost'].fillna(value=dataset['Estimated Cost'].mean(), inplace=True)
dataset['Revised Cost'].fillna(value=dataset['Revised Cost'].mean(), inplace=True)
dataset['Plansets'].fillna(value=dataset['Plansets'].mean(), inplace=True)
dataset['Existing Units'].fillna(value=dataset['Existing Units'].mean(), inplace=True)
dataset.isnull().sum()*100/dataset.isnull().count()
```

```python
# Following features are replaced with the value of maximum occurance (75%)
    # Existing Construction Type - 5          
    # Proposed Construction Type - 5
    # Proposed Units
dataset['Existing Construction Type'] = dataset['Existing Construction Type'].replace(np.nan, 5)
dataset['Proposed Construction Type'] = dataset['Proposed Construction Type'].replace(np.nan, 5)
dataset['Proposed Units'] = dataset['Proposed Units'].replace(np.nan, 5)
dataset.isnull().sum()*100/dataset.isnull().count()
```

```python
# Following features are replaced with the top value
    # Existing Use - 1 family dwelling
    # Proposed Use - 1 family dwelling
    # Existing Construction Type Description  - wood frame (5)     
    # Proposed Construction Type Description - wood frame (5)
# dataset['Completed Date'] = dataset['Completed Date'].replace(np.nan, year=2015, month=10, day=30)
dataset['Existing Use'] = dataset['Existing Use'].replace(np.nan, '1 family dwelling')
dataset['Proposed Use'] = dataset['Proposed Use'].replace(np.nan, '1 family dwelling')
dataset['Existing Construction Type Description'] = dataset['Existing Construction Type Description'].replace(np.nan, 'wood frame (5)')
dataset['Proposed Construction Type Description'] = dataset['Proposed Construction Type Description'].replace(np.nan, 'wood frame (5)')
dataset.isnull().sum()*100/dataset.isnull().count()
```

```python
# As following values cannot predicted and number of NaN values are less; raws are deleted
    # Street Suffix
    # Description
    # Issued Date
    # Completed Date
    # First Construction Document Date
    # Permit Expiration Date
    # Supervisor District
    # Neighborhoods - Analysis Boundaries
    # Zipcode
    # Location
dataset.dropna(subset=['Street Suffix','Description','Completed Date','Issued Date','First Construction Document Date','Permit Expiration Date','Supervisor District','Neighborhoods - Analysis Boundaries','Zipcode','Location'], inplace=True)
dataset.isnull().sum()*100/dataset.isnull().count()
```

```python
# Following coloms are deleted as they not provide usefull info and number of NaN are high
    # Street Number Suffix
    # Unit
    # Unit Suffix

dataset.dropna(axis=1, inplace=True)
dataset.isnull().sum()*100/dataset.isnull().count()
```
